```python
import pandas as pd
import datetime as dt
```

```python
weather = pd.read_csv('weather_alerts_grouped.csv')
weather['END_DATETIME'] = pd.to_datetime(weather['END_DATETIME'])
weather['BEGIN_DATETIME'] = pd.to_datetime(weather['BEGIN_DATETIME'])
```

```python
crime = pd.read_csv('PropertyCrimes_CLEAN.csv')
crime['Incident_datetime'] = pd.to_datetime(crime['Incident_datetime'])
```

```python
#extract date from datetime for both crimes and weather
weather['DATE'] = weather['BEGIN_DATETIME'].dt.date
crime['DATE'] = crime['Incident_datetime'].dt.date
```

```python
crime_weather = pd.merge(crime, weather, on='DATE', how='left')
crime_weather[crime_weather['EVENT_TYPE'].notnull()]
```

```python
crime_weather
```

```python
#check if the Incident_datetime is within the weather alert window
crime_weather['in_window'] = (crime_weather['Incident_datetime'] > crime_weather['BEGIN_DATETIME'])
crime_weather[crime_weather['in_window']]
```

```python
crime_weather.columns
```

```python
crime_weather.info()
```

```python
crime_weather.drop(['date','date.1','EVENT_TYPE_col','Incident_datetime','BEGIN_DATETIME','END_DATETIME'], axis=1, inplace=True)
crime_weather.info()
```

```python
crime_weather.nunique()
```

```python
crime_weather.drop_duplicates(inplace=True)
```

```python
crime_weather.info()
```

```python
#if in_window is False, then set EVENT_TYPE to null
crime_weather.loc[crime_weather['in_window'] == False, 'EVENT_TYPE'] = None
```

```python
#drop in_window column
crime_weather.drop(['in_window'], axis=1, inplace=True)
crime_weather
```

```python
crime_no_weather = crime_weather[crime_weather['EVENT_TYPE'].isnull()].copy()
```

```python
crime_extreme_weather = crime_weather[crime_weather['EVENT_TYPE'].notnull()].copy()
```

```python
#export to csv in Data_files folder
crime_weather.to_csv('crime_weather_all_CLEANED.csv')
crime_no_weather.to_csv('crime_no_weather_CLEANED.csv')
crime_extreme_weather.to_csv('crime_extreme_weather_CLEANED.csv')
```
