```python
import pandas as pd
```

```python
df_incidents11 = pd.read_csv('2011_Incident (1).csv')
df_incidents12 = pd.read_csv('2012_Incident.csv')
df_incidents13 = pd.read_csv('2013_Incident.csv')
df_incidents14 = pd.read_csv('2014_Incident.csv')
df_incidents15 = pd.read_csv('2015_Incident.csv')
df_incidents16 = pd.read_csv('2016_Incident.csv')
```

```python
df_property11 = pd.read_csv('2011_Property.csv')


df_property12 = pd.read_csv('2012_Property.csv')


df_property13 = pd.read_csv('2013_Property.csv')


df_property14 = pd.read_csv('2014_Property.csv')


df_property15 = pd.read_csv('2015_Property.csv')


df_property16 = pd.read_csv('2016_Property.csv')
```

```python
combinedincidents_df = pd.concat([df_incidents11, df_incidents12, df_incidents13, df_incidents14, df_incidents15, df_incidents16 ], ignore_index=True)
combinedproperty_df = pd.concat([df_property11, df_property12, df_property13, df_property14, df_property15, df_property16 ], ignore_index=True)

combined_all = combinedincidents_df.merge(combinedproperty_df, on='Complaint_No', how='left')

combined_all
```

```python
combined_all.columns
```

```python
columns_to_keep = ['Complaint_No','NIBRS_Hi_Class','Location_Type','Incident_From_Date',' Incident_From_Time', 'Zipcode', 'Property_No', 'Property_type', 'Property_Desc']

combined_all = combined_all[columns_to_keep]
combined_all
```

```python
unique_values = combined_all['NIBRS_Hi_Class'].unique()
print( unique_values)
```

```python
#Cleaning white spaces
combined_all['NIBRS_Hi_Class'] = combined_all['NIBRS_Hi_Class'].str.strip()

unique_values = combined_all['NIBRS_Hi_Class'].unique()
print( unique_values)
```

```python

# List of property crimes to keep
property_crimes_to_keep = [
    'Burglary/B&E', 'Theft', 'Motor Vehicle Theft', 'Theft from Motor Vehicle',
    'All other Thefts', 'Theft of Motor Vehicle Parts from Vehicle',
    'Theft from Building', 'Theft from Coin Operated Machinery', 'Arson',
    'Damage/ Vandalism to Property', 'Shoplifting', 'Embezzlement'
]

# Filter the dataframe for the specified property crimes
filtered_df = combined_all[combined_all['NIBRS_Hi_Class'].isin(property_crimes_to_keep)]
```
