```python
combined_df.columns
```

```python
# Combine "Victim_Race" and "Victim_Ethnicity" columns into a new column "Victim_Ethnicity"
combined_df['Victim_Ethnic'] = combined_df['Victim_Race'] + ' ' + combined_df['Victim_Ethnicity']
combined_df = combined_df.drop(['Victim_Race', 'Victim_Ethnicity'], axis=1)
combined_df.columns
```

```python
combined_df.columns.str.strip()
```

```python
columns_to_keep = ['Complaint_No','NIBRS_Hi_Class','Location_Type','Incident_From_Date',' Incident_From_Time', 'Zipcode', 'Property_No', 'Property_type', 'Property_Desc', '2010 Pop', 'Median Household Income', 'Victim_Age', 'Victim_Gender', 'Victim_Ethnic']

combined_df = combined_df[columns_to_keep]
combined_df
```

```python
combined_df = combined_df.rename(columns={' Incident_From_Time':'Incident_Time'})
combined_df = combined_df.rename(columns={'Incident_From_Date':'Incident_Date'})
combined_df = combined_df.rename(columns={'NIBRS_Hi_Class':'Property_Crimes'})

combined_df
```

```python
for col in combined_df.columns:
    if combined_df[col].dtype == 'object':
        combined_df[col] = combined_df[col].str.strip()
```

```python
combined_df.info()
```

```python
combined_df['Incident_Date_datetime'] = pd.to_datetime(combined_df['Incident_Date'], errors='coerce')
```

```python
combined_df['Incident_Time'] = combined_df['Incident_Time'].astype(str)
combined_df['Incident_Time'] = combined_df['Incident_Time'].str.strip()
combined_df['Incident_Time'] = combined_df['Incident_Time'].apply(lambda x: '0'+x if len(x) == 3 else x)
combined_df['Incident_Time'] = combined_df['Incident_Time'].apply(lambda x: '00'+x if len(x) == 2 else x)
combined_df['Incident_Time'] = combined_df['Incident_Time'].apply(lambda x: '000'+x if len(x) == 1 else x)
combined_df['Incident_Time'] = pd.to_numeric(combined_df['Incident_Time'], errors='coerce')
```

```python
combined_df['Incident_Time_time'] = pd.to_datetime(combined_df['Incident_Time'],format= '%H%M',errors='coerce').dt.time
combined_df.info()
```

```python
combined_df = combined_df.dropna(subset=['Incident_Date_datetime','Incident_Time_time'])
combined_df.info()
```

```python
combined_df['Incident_datetime'] = combined_df.apply(lambda x: pd.datetime.combine(x['Incident_Date_datetime'],x["Incident_Time_time"]), axis = 1)
combined_df
```

```python
combined_df.drop(['Incident_Date_datetime','Incident_Time_time','Incident_Date','Incident_Time'], axis=1, inplace=True)
```

```python
combined_df.info()
```

```python
# List of property crimes to keep
property_crimes_to_keep = [
    'Burglary/B&E', 'Theft', 'Motor Vehicle Theft', 'Theft from Motor Vehicle',
    'All other Thefts', 'Theft of Motor Vehicle Parts from Vehicle',
    'Theft from Building', 'Theft from Coin Operated Machinery', 'Arson',
    'Damage/ Vandalism to Property', 'Shoplifting', 'Embezzlement'
]

Property_df = combined_df[combined_df['Property_Crimes'].isin(property_crimes_to_keep)]
Property_df
```

```python
Property_df['Median Household Income'] = Property_df['Median Household Income'].str.replace(',', '').str.replace('$', '').astype(float)
Property_df
```
